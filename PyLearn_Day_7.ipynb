```python
#Hangman Game part 1

import random

world_list=['ardvark','baboon','camel']

#To-Do-1 --It is important to remember the syntax of random list
chosen_word=random.choice(world_list)
print(chosen_word)

chosen_word_list=list(chosen_word)
print(f'The list view is chosen_word_list {chosen_word_list}')

#To-Do-2
guess=''
guessed_letter=input('Guess a letter for the chosen word: ')
guess=guessed_letter.lower()

#To-Do-3
word_len=len(chosen_word)
flag=0
for i in chosen_word_list:
    if i==guess:
        flag=1
        print('Right')
    else:
        print('Wrong')
        
if flag==1:
    print('You guessed correct letter')
else:
    print('Over-all Wrong guess')



```

```python
#Hangman Game trial 3
world_list=['ardvark','baboon','camel']
import random

guess_counter=0

chosen_word=random.choice(world_list)
word_len=len(chosen_word)
print(f'The chosen word is {chosen_word}')

guess=input('Guess a letter: ').lower()


#To-Do-1 create an empty list with same number of spaces
created_null_list=[]
for i in range(0,word_len):
    created_null_list.append('_')

print(f'The null list is {created_null_list}')

#To-Do-2
index=0
hang=0

for i in chosen_word:
    if i==guess:
        created_null_list[index]=guess
    else:
        created_null_list[index]='_'
    index +=1
    hang +=1

print(f'The updated list is {created_null_list}')

#To-Do-3
made_word=''.join(created_null_list)
print(f'The composed word is {made_word} ')

```

```python
#Hangman Game trial 4
world_list=['ardvark','baboon','camel']
import random

guess_counter=0
lives_counter=6

chosen_word=random.choice(world_list)
word_len=len(chosen_word)
print(f'The chosen word is {chosen_word}')



#To-Do-1 create an empty list with same number of spaces
display=[]
for i in range(0,word_len):
    display +='_'
    #display.append('_')
    

print(f'Fill in the blanks {display}')


#To-Do-2

#Loop addition
end_of_game=False

while not end_of_game:
    
    guess=input('Guess a letter: ').lower()

            
    if guess in display:
        print(f'You have already guessed {guess}')
    
    if guess not in chosen_word:
        print('You have chosed a wrong letter, you lose a life')

    for position in range(word_len):
        letter=chosen_word[position]
        if letter==guess:
            display[position]=letter
            
            

    print(f'The updated list is {display}')

    #To-Do-3
    made_word=''.join(display)
    print(f'The composed word is {made_word} ')
    
    
    if '_' not in display:
        end_of_game=True
        print('You Win')
        
    if guess not in chosen_word:
        lives_counter -=1
        if lives_counter==0:
            end_of_game=True
            print('You Lose')

```

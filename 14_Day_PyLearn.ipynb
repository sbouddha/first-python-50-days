```python
def my_func():
    x = 10    # local variable
    print("Inside my_func, x =", x)

x = 5    # global variable
my_func()
print("Outside my_func, x =", x)
```

```python
x = 10   # global variable

def my_func():
    global x
    x = 5    # global variable
    print("Inside my_func, x =", x)

my_func()
print("Outside my_func, x =", x)
##This is very important and useful
```

```python
#Program: Higher or Lower
import random
import time

from IPython.display import clear_output

def get_2_options():
    global dict_A,dict_B,name_A,desc_A,country_A,follower_count_A,name_B,desc_B,country_B,follower_count_B
    
    dict_A, dict_B = random.sample(game_data,2)

    name_A=dict_A.get('name')
    desc_A=dict_A.get('description')
    country_A=dict_A.get('country')
    follower_count_A=dict_A.get('follower_count')

    name_B=dict_B.get('name')
    desc_B=dict_B.get('description')
    country_B=dict_B.get('country')
    follower_count_B=dict_B.get('follower_count')
    print('You are playing Higher or Lower, Below are your 2 options: ')
    print(f'Compare A : {name_A}, a {desc_A}, from {country_A} \nvs\nAgainst B : {name_B}, a {desc_B}, from {country_B}')


def is_higher():
    if follower_count_A > follower_count_B:
        return  'A'
    else:
        'B'
        
def user_guess():
    global user_input
    user_input=input('Who has more followers? Type "A" or "B": ').upper() 
    return user_input

def add_score():
    global final_score
    final_score= final_score + 1
    return final_score

final_score=0
def game_higher_lower():
    get_2_options()
    
    if user_guess()==is_higher():
        print('Yay !!! You got that Correct, new challenge coming in a second')
        add_score()
        time.sleep(2)  # Wait for 1 second
        clear_output(wait=True)
        game_higher_lower()
    else:
        print(f'Sorry, that is wrong. Final score: {final_score}')


game_higher_lower()



```
